```python
import pandas as pd
import folium
import seaborn as sns
import matplotlib.pyplot as plt
```

# Densite 5km sur 5km

```python
coeff_mul = 1

state_geo = "data/geospace5km.json"
state_data = pd.read_csv("data/densite5km.csv", index_col = 0)
state_data.value = state_data.value.apply(str)
state_data.value = state_data.value.astype(str)
state_data = state_data.drop(columns = ["x","y","nblogement"])

state_data.densite =state_data.densite * coeff_mul


m = folium.Map(location=[47.8445, 7.2139], zoom_start=8)

info_maison = pd.read_csv("data/infoMaison.csv",index_col = 0)

tooltip = "Click me!"
my_formatter = "{0:.2f}"
for row in info_maison.itertuples():
    folium.Marker(
        [row.latitude,row.longitude], popup=f"<i> DPE : {my_formatter.format(row.dpe)} \n {row.adresse} {row.ville} {row.code_postale}\n {row.type_maison}</i>", tooltip=tooltip
    ).add_to(m)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["value", "densite"],
    key_on="feature.id",
    bins = 9,
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Densit de logement/km",
    
).add_to(m)

folium.LayerControl().add_to(m)
m
```
